```python
import math

for name in dir(math):
    print(name, end='\t')
    
```

```python
# ceil(), floor(), trunc()
print(math.ceil(3.6))
print(math.floor(3.6))
print(math.trunc(3.6))
```

```python
#ceil()
print(math.ceil(3.6))
print(math.ceil(3.99))
print(math.ceil(3.234))
print(math.ceil(3.2))
print(math.ceil(3.0))
print(math.ceil(-3.6))
print(math.ceil(-3.1))
```

```python
#floor()
print(math.floor(3.6))
print(math.floor(3.99))
print(math.floor(3.234))
print(math.floor(3.2))
print(math.floor(3.0))
print(math.floor(-3.6))
print(math.floor(-3.1))
```

```python
#trunc
print(math.trunc(3.6))
print(math.trunc(3.99))
print(math.trunc(3.234))
print(math.trunc(3.2))
print(math.trunc(3.0))
print(math.trunc(-3.6))
print(math.trunc(-3.1))
```

```python
#factorial()
# 3! = 3*2*1 = 6
# 4! = 4*3*2*1 = 24
print(math.factorial(3))
print(math.factorial(4))
```

```python
#sqrt() alway returt float
print(math.sqrt(16))
print(math.sqrt(100))
```

```python
#hypot()
print(math.hypot(3,4))
```
