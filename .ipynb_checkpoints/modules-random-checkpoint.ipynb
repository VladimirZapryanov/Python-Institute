```python
import random

def generate_tickets(ticket_count, max_number):
    random_numbers = []
    
    for i in range(max_number):
        random_numbers.append(i)
    
    ticket_numbers = random.sample(random_numbers, ticket_count)
    win_number = random.sample(ticket_numbers, 1)
        
    return(ticket_numbers, int(win_number[0]))
    
print(generate_tickets(5, 10))
```
